```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from wordcloud import WordCloud

datafile = Path('Atlanta.xlsx')
df = pd.read_excel(datafile)

"""Bar for the IMDB ratings"""

plt.bar(df['Episode Name'],df['IMDB Rating'],label='Rating')
plt.ylabel('IMDB')
plt.xlabel('Episode')
plt.xticks(rotation='vertical')
plt.title(' IMDB rating')

plt.legend()
plt.savefig('HistImdbEpisode.png')
plt.show()


```

```python
"""IMDB ratings of the two seasons"""

df['IMDB Rating'].iloc[0:12].plot(color='red',marker='o',label='Season 1')
df['IMDB Rating'].iloc[10:22].plot(color='black',marker='o',label='Season 2')
plt.ylabel('IMDB')
plt.xlabel('Episode')
plt.title('IMDB ratings per episode')

plt.legend()
plt.savefig('LineImdbSeasons.png')
plt.show
```

```python
"""Which director has directed the most number of episodes?"""

df['Directors'].value_counts()[:4].plot(kind='barh')
plt.ylabel('Directors')
plt.xlabel('Amount')
plt.title('Amount of directed episodes')
plt.savefig('HistDirected.png')
plt.show


```

```python
from collections import Counter

"""Most Frequent Words"""

decription = df['Description']
split_it = str(decription).split()
Counter = Counter(split_it)

wordcloud = WordCloud()
wordcloud.generate_from_frequencies(Counter)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('WCFrequincies.png')
plt.show()

```
